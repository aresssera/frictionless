```
frame = pd.read_csv('ogd102_fuellstand_gasspeicher.csv')
frame
```

```
# convert dictionary schema into frictionless schema object
schema = Schema({'fields': [{'name': 'Datum', 'type': 'date', 'format': 'default', 'title@de': 'Datum', 'title@fr': '', 'description@de': '', 'description@fr': ''}, {'name': 'Speicherregion', 'type': 'string', 'format': 'default', 'title@de': 'Speicherregion', 'title@fr': '', 'description@de': '', 'description@fr': ''}, {'name': 'Speicherstand_TWh', 'type': 'number', 'format': 'default', 'title@de': 'Speicherstand_TWh', 'title@fr': '', 'description@de': '', 'description@fr': ''}, {'name': 'Lagerkapazitaet_prozent', 'type': 'number', 'format': 'default', 'title@de': 'Lagerkapazitaet_prozent', 'title@fr': '', 'description@de': '', 'description@fr': ''}]})

# perform validation using schema matched to uploaded file
report = validate(frame, schema=schema)
report
```

```
text = ''

for err in report.tasks[0].errors:
  text = text + err.title + '\n' + err.message + '\n\n'
```
